```python
#Gerando array de seeds

def GenerateSeed(seed_parameter):
  L = []
  rand.seed(seed_parameter)
  for i in range (0,2000):
    L.append(rand.random())
  return L

## Definição do cliente
class Cliente:

    def __init__(self, TempRelogioAnterior, TempIniAnt, TSAnt, TempoFimAnt, seedTEC, seedTS):
        rand.seed(seedTEC)
        self.TEC = -2 * math.log(rand.choice(GenerateSeed(seedTEC)))
        rand.seed(seedTS)
        self.TS = -3 * math.log(rand.choice(GenerateSeed(seedTS)))
        self.Tempo_Relogio = self.TEC + TempRelogioAnterior   #TEC + Tempo do relogio do cliente anterior
        if((TempIniAnt + TSAnt) < self.Tempo_Relogio):
          self.Tempo_inicio = self.Tempo_Relogio
        else:
          self.Tempo_inicio = TempIniAnt + TSAnt
        self.Tempo_fim = self.Tempo_inicio + self.TS
        self.Tempo_fila = self.Tempo_inicio - self.Tempo_Relogio
        self.Tempo_sistema = self.TS + + self.Tempo_fila
        self.Tempo_livre_caixa = self.Tempo_inicio - TempoFimAnt
   
  
    
```

# Criação do sistema de filas para N Filas e N Caixas

```python
# Calculando Filas de clientes
# Neste Caso cada fila tem um caixa, ou seja N_Filas = Número de Caixas, pois cada caixa tem uma fila

# Preencher essa função da seguinte forma: 
#(N_Filas(Número de caixas), Num_Clientes(Número de Clientes a serem atendidos), seedTEC(seed escolhida pelo usuário para o sistema inteiro), seedTS(seed escolhida pelo usuário para o sistema inteiro))

def CriaSistemasFila(N_Filas,Num_Clientes,seedTEC,seedTS):
  Filas = []
  seedNFila = []
  for x in range(0, N_Filas):
    if(x == 0):
      y = rand.random()
      Filas.append([Cliente(0,0,0,0, seedTEC*y*100, seedTS*y*200)])
      seedNFila.append([seedTEC*y*100, seedTS*y*200])
    else:
      Filas.append([Cliente(0,0,0,0, seedTEC*x*100, seedTS*x*200)])   #CRIANDO GRUPO DE FILAS, CADA FILA TEM UM CAIXA
      seedNFila.append([seedTEC*x*100, seedTS*x*200])
  for i in range(0, Num_Clientes - N_Filas):
    menorTempo = math.inf                                             ## Menor tempo de final de serviço entre todas as filas antes de inserir o próximo cliente
    proximaFila = -1                                                  ## Proxima fila a receber um cliente
    for x in range(0, N_Filas):
      if(Filas[x][-1].Tempo_fim < menorTempo):
        menorTempo = Filas[x][-1].Tempo_fim
        proximaFila = x
    Filas[proximaFila].append(Cliente(Filas[proximaFila][-1].Tempo_Relogio, Filas[proximaFila][-1].Tempo_inicio, Filas[proximaFila][-1].TS, 
                                      Filas[proximaFila][-1].Tempo_fim, seedNFila[proximaFila][0]*i*100, seedNFila[proximaFila][1]*i*200)) 
    
  return Filas

# Neste caso teste temos 10 filas para 100 clientes.
sistemaFila = CriaSistemasFila(5, 100, 3000, 2000)

#Printando Cada Fila de clientes
count = 1
print("####   Sistema EXEMPLO ##### \n\n\n")
for Fila in sistemaFila:
  print("Fila: ",count,"\n")
  count += 1
  for i in range(0,len(Fila)):
    print("Cliente", i+1 ,":","%.2f" %Fila[i].TEC, "%.2f" %Fila[i].TS, "%.2f" %Fila[i].Tempo_Relogio,"%.2f" %Fila[i].Tempo_inicio,"%.2f" %Fila[i].Tempo_fim, "%.2f" 
    %Fila[i].Tempo_fila, "%.2f" %Fila[i].Tempo_sistema, "%.2f" %Fila[i].Tempo_livre_caixa, "\n")



# Agora vamos fazer uma simulação para vários sistemas cada um com num_Filas (Número de Caixas)

num_Sistemas = 500  # Numero de sistemas na simulação
num_Filas = 2      # Numero de Filas (Caixas) em cada sistema
num_Clientes = 100    # Número de clientes em cada sistema
Sistemas = []

for j in range(0, num_Sistemas):
  Sistemas.append(CriaSistemasFila(num_Filas, num_Clientes, num_Sistemas*100*230, num_Sistemas*1000*3000))

```

# Médias das filas do sistema sistema para N sistemas

```python
## Calculando Médias de espera na Fila

def TempoMedioDeEsperaNasFilas(Filas):

  SomaMediaFilas = 0
  SomaTempoCadaFila = 0
  for Fila in Filas:
    for Cliente in Fila:
      SomaTempoCadaFila += Cliente.Tempo_fila
    MediaNFila = SomaTempoCadaFila/len(Fila)
    SomaTempoCadaFila = 0
    SomaMediaFilas += MediaNFila

  tempoMedio = SomaMediaFilas / len(Filas)
  return tempoMedio

SomaMedias = 0
for Sistema in Sistemas:
  SomaMedias += TempoMedioDeEsperaNasFilas(Sistema)

print("Tempo de espera Médio no sistema:", SomaMedias/len(Sistemas))

## Fim cálculo de médias de espera na fila

## Tempo médio Gasto no sistema

def TempoMedioGastoSistema(Filas):

  SomaMediaFilas = 0
  SomaTempoCadaFila = 0
  for Fila in Filas:
    for Cliente in Fila:
      SomaTempoCadaFila += Cliente.Tempo_sistema
    MediaNFila = SomaTempoCadaFila/len(Fila)
    SomaTempoCadaFila = 0
    SomaMediaFilas += MediaNFila

  tempoMedio = SomaMediaFilas / len(Filas)
  return tempoMedio

SomaMedias = 0
for Sistema in Sistemas:
  SomaMedias += TempoMedioGastoSistema(Sistema)

print("\nTempo Médio gasto no sistema:", SomaMedias/len(Sistemas))









## Fim tempo médio gasto no sistema
```
